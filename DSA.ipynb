```python
# Print the number of set bits for the given number
# 16 -> 10000 -> 1
# 15 -> 01111 -> 4
# & - and
# | - or
# >> - right shift
# << - left shift

def  countSetBits(n):
    count = 0
    while (n):
        count += n & 1
        n >>= 1
    return count

countSetBits(int(input("Enter the number: ")))
```

```python
# Find factorial of a number using recursion.
# 1! = 1
# 2! = 1 * 2 = 2
# 3! = 2 * 3 = 6
# 4! = 6 * 4 = 24
# 5! = 24 * 5 = 120

def fact(n):
  if n == 1:
    return 1
  return n * fact(n-1)

n = int(input("Enter the number: "))
if n < 0:
  print("Factorial does not exist for negative numbers")
elif n == 0:
  print("The factorial of 0 is 1")
else:
  print("The factorial of {} is {}".format(n,fact(n)))
```

```python
# https://leetcode.com/problems/find-minimum-operations-to-make-all-elements-divisible-by-three/description/
class Solution:
    def minimumOperations(self, nums: List[int]) -> int:
        result = [1 if i%3 != 0 else 0 for i in nums]
        return sum(result)
```

```python
# https://leetcode.com/problems/shuffle-the-array/description/
class Solution:
    def shuffle(self, nums: List[int], n: int) -> List[int]:
        result = []
        for i in range(n):
            result += [nums[i], nums[i+n]]
        return result
```

```python
# https://leetcode.com/problems/number-of-good-pairs/
class Solution:
    def numIdenticalPairs(self, nums: List[int]) -> int:
        pair_count = 0
        for i in range(len(nums)):
            pair_count += nums[i+1:].count(nums[i])
        return pair_count
```

```python
# https://leetcode.com/problems/divisible-and-non-divisible-sums-difference/
class Solution:
    def differenceOfSums(self, n: int, m: int) -> int:
        x = n // m
        total = (n * (n + 1))//2
        n2 = (x * (x + 1))//2 * m
        return (total - (2 * n2))
```

```python
# https://leetcode.com/problems/two-sum/
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for i in range(len(nums)):
            x = target - nums[i]
            if x in nums[i+1:]:
                li = [i]
                li.append(nums[i+1:].index(x) + i + 1)
                return li
```

```python
# https://leetcode.com/problems/find-words-containing-character/
class Solution:
    def findWordsContaining(self, words: List[str], x: str) -> List[int]:
        result = []
        for i in range(len(words)):
            if x in words[i]:
                result.append(i)
        return result
```

```python
# https://leetcode.com/problems/richest-customer-wealth/
class Solution:
    def maximumWealth(self, accounts: List[List[int]]) -> int:
        richest = 0
        for i in accounts:
            temp = 0
            for j in i:
                temp += j
            if temp > richest:
                richest = temp
        return richest
```

```python
# Algorithm for merge sort
merge_sort(arr, low , high):
  if low < high:
    mid = (low + high) // 2
    merge_sort(arr, low, mid)
    merge_sort(arr, mid+1, high)
    merge(arr, low, mid, high)


merge(arr, low, mid, high):
  B = [0 for i in range(high - low + 1)]
  i = low
  j = mid + 1
  k = 0
  while i <= mid and j <= high:
    if arr[i] < arr[j]:
      B[k] = arr[i]
      i += 1
    else:
      B[k] = arr[j]
      j += 1
    k += 1
  while i <= mid:
    B[k] = arr[i]
    i += 1
    k += 1
  while j <= high:
    B[k] = arr[j]
    j += 1
    k += 1
  for i in range(low, high + 1):
    arr[i] = B[i - low]
  return arr
```

```python
# Quick Sort Algorithm:
quick_sort(arr, low, high):
  if low < high:
    p = partition(arr, low, high)
    quick_sort(arr, low, p - 1)
    quick_sort(arr, p + 1, high)


partition(arr, low, high):
  pivot = arr[high]
  i = m + 1
  j = low
  while i <= j:
    while arr[i] <= pivot and i < j:
      i += 1
    while arr[j] > pivot and i < j:
      j -= 1
    if i < j:
      temp = arr[i]
      arr[i] = arr[j]
      arr[j] = temp
    temp = arr[j]
    arr[j] = arr[high]
    arr[high] = temp
  return j
```
